#Example-1
Your first task is to simulate and plot the correlated time series

                    x[t+1] =  x[t] + e_v[t+1],     t=0,...,T, x0=0, T=200, =0.9

The sequence of shocks {e_v} is assumed to be IID and standard normal.

```python
def example1(,T,l):
    e_v=np.random.randn(T+1)
    x=np.empty(T+1)
    x[0]=0

    for t in range(T):
        x[t+1]=*x[t]+e_v[t+1]

    plt.plot(x,label=l)

example1(0.9,200,'balance')
plt.legend()
plt.show()
```

#Example-3
Similar to the previous exercises, plot the time series

    x[t+1] =  |x[t]| + e_v[t+1],     t=0,...,T, x0=0, T=200, =0.9

```python
def absolexample1(,T,l):
    e_v=np.random.randn(T+1)
    x=np.empty(T+1)
    x[0]=0

    for t in range(T):
        x[t+1]=*np.abs(x[t])+e_v[t+1]

    plt.plot(x,label=l)

absolexample1(0.9,200,'balance')
plt.legend()
plt.show()
```

```python
def fact(n):
    if n==1:
        return n
    return n*fact(n-1)

print(fact(3))
```

```python
def flip(k):
    count=0
    bill=0
    for _ in range(10):
        a=np.random.choice(['H','T'])
        print(a)
        if a=='H':
            count+=1
            if count>=k:
                bill=1
        else:
            count=0
    return bill

flip(3)
```

```python
def fib(t):
    if t==0:
        return 0
    if t==1:
        return 1
    return fib(t-1)+fib(t-2)

fib(7)
```
